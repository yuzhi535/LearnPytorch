```python
import torch
import torchvision
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import torch.nn as nn
import torchvision.transforms as transforms
from torchvision import datasets
```

```python
transform = transforms.Compose(
    [transforms.ToTensor(), transforms.CenterCrop(128), transforms.Normalize(mean=.5, std=.5)])

dataset = datasets.ImageFolder(root='./data/dog_cat/', transform=transform)

example = iter(dataset)

sample, lables = example.__next__()
print(sample.shape)
```

```python
example.__next__()
sample, lables = example.__next__()
plt.imshow(np.transpose(sample, (1, 2, 0)))
plt.show()
```

```python
data_loader = torch.utils.data.DataLoader(dataset=dataset, batch_size=4, shuffle=True, num_workers=0)
```

```python
train_features, train_labels = next(iter(data_loader))

for i in range(4):
    image = train_features[i]
    plt.subplot(2, 2, i+1)
    image = np.transpose(image.numpy(), (1, 2, 0))
    plt.imshow(image)
plt.show()
```
