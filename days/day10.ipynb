```python
import torch
import torchvision
import cv2 as cv
import matplotlib.pyplot as plt
from torchvision import transforms, datasets
import torch.nn as nn
import numpy as np
```

```python
T = transforms.Compose([transforms.CenterCrop(50), transforms.ToTensor()])
dataset = datasets.CIFAR10(root='../data', train=True, transform=Tï¼Œ download=True)
```

```python
batch_size=4

dataloader = torch.utils.data.DataLoader(dataset=dataset, shuffle=True, batch_size=4)
```

```python
examples = iter(dataloader)
```

```python
samples, labels = next(examples)
samples.shape, labels.shape
```

```python
for i in range(4):
    image = samples[i]
    image = np.transpose(image, (1,2,0))
    image = image.numpy()
    image = cv.cvtColor(image, cv.COLOR_RGB2GRAY)
    image = np.uint8(image*255)
    image = cv.applyColorMap(image, cv.COLORMAP_JET)
    plt.subplot(1,4,i+1)
    plt.imshow(image, cmap='gray')
plt.show()
```
