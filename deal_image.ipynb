# show images

```python
import torch
import torchvision
import matplotlib.pyplot as plt
import numpy as np
import torchvision.transforms as transforms
from PIL import Image
import os
```

```python
image_dir = './images'
images = os.listdir(image_dir)
images_num = len(images)

t1 = transforms.Compose([transforms.ToTensor(), transforms.Resize((255, 255)), transforms.RandomHorizontalFlip(), transforms.RandomVerticalFlip()])

images = [ Image.open(os.path.join(image_dir, images[i])) for i in range(images_num)]

images = [ t1(i)  for i in images]

image = torchvision.utils.make_grid(images)

image = np.transpose(image, (1, 2, 0))
plt.imshow(image)
plt.show()
```

```python
t2 = transforms.Compose([transforms.ToTensor()])
dataset = dataset = torchvision.datasets.CIFAR10(root='./data', train=True, download=False, transform=t2)
data = torch.utils.data.DataLoader(dataset=dataset, shuffle=True, batch_size=4)

example = iter(data)
```

```python

sample, label = example.next()
print(sample.shape, label.shape)
```

```python
for i in range(4):
    plt.subplot(2, 2, i + 1)
    image = sample[i]
    plt.imshow(np.transpose(image.numpy(), (1, 2, 0)))
plt.show()
```
